# All Linked list Ops

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None 
```

```python
class Linkedlist:
    def __init__(self):
        self.head = None 
    
    def print(self):
        temp = self.head 
        while temp is not None:
            print(temp.data,end="->")
            temp = temp.next 
        print()
    def InsertfromHead(self,newValue):
        newNode = Node(newValue)
        newNode.next = self.head
        self.head = newNode
    
    def InsertfromTail(self,newValue):
        newNode = Node(newValue)
        if self.head is None:
            self.head = newNode
            return 

        last = self.head 
        while last.next:
            last =last.next
        last.next = newNode 


    def Insertbetween2Node(self,middlevalue,newValue):
        if middlevalue is None:
            print("the mentioned node is absent :")
            return 
        newNode = Node(newValue)
        newNode.next = middlevalue.next 
        middlevalue.next = newNode
    
    def remove(self,value):
        head = self.head        
        # Yeah specifically code hai if 0th index is the value that 
        # we wanna delete     
        if head is not None: 
            if head.data == value:
                self.head = head.next 
                head = None
            if head == None:
                return             
            # This code is for if random middle element delete krna hai 
            while(head is not None):
                if head.data == value:
                    break 
                previous = head 
                head = head.next               
            previous.next = head.next  # prevous tha value ka jo uska agla hojayega humare value ka agla ishse value beechmai margaya
            head = None # now making it None 
          
```

```python
ll = Linkedlist() 
firstNode = Node(10)
secondNode = Node(90)
thirdNode = Node(99)

ll.head = firstNode
firstNode.next = secondNode
secondNode.next = thirdNode

ll.print()

print("-----InsertfromHead Operation------")

ll.InsertfromHead("fromHead")
ll.print()

print("-----InsertfromTail Operation------")

ll.InsertfromTail(909)
ll.print()

print("-----Insertbetween2Node Operation------")

ll.Insertbetween2Node(ll.head.next.next,"inbetween")
ll.print()

print("-----Remove Operation------")


ll.remove("inbetween")
ll.print()


ll.remove("fromHead")
ll.print()


ll.remove(10)
ll.print()
```
