### Array Implementation of Queue

```python
class Queue:
    def __init__(self):
        self.queue = []

    def check_empty(self):
        if len(self.queue)<1:
            return True
        else:
            return False 
    
    def enqueue(self,item):
        self.queue.append(item)
    
    def dequeue(self):
        # if len(self.queue) < 1:
        #     None
        # return self.queue.pop(0)
        if self.check_empty()==True:
            return "Queue is empty"
        else:
            self.queue.pop(0)

    def display(self):
        print(self.queue)
```

```python
Q = Queue()

print("Queue at starting :",Q.display())

Q.enqueue(99)
Q.enqueue(999)
Q.enqueue(0)
Q.enqueue(1)
Q.display()
Q.dequeue()
Q.display()

Q.dequeue()
Q.display()
```
