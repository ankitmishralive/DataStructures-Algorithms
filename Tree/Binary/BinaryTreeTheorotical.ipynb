# *Binary Tree* 

A binary tree is a tree data structure in which each parent node can have at most two children.   
Each node of a binary tree consists of three items

- data item

- address of left child

- address of right child

## *Types of Binary Tree* 

- 1.Full binary tree   

A full binary tree is a special type of binary tree in which every parent node has either 2 or no children 

- 2.Perfect Binary Tree 

A perfect binary tree is a type of binary tree in which every internal node has exactly 2 child nodes & all the leaf nodes are at same level. 

- 3.Complete Binary Tree 

A complete binary tree is just like a full binary tree, but with 2 major differences 

1. Every level must be completely filled. 
2. All the leaf elements must lean towards the left. 
3. The last leaf element might not have a right siblings ie. a complete binary tree doesn't have to be a full binary tree.



# *Difference between General Tree & Binary Tree*


In binary tree a node can have atmost 2 children for a particular node.   

In General tree is a tree in which each node can have either zero or many child nodes. It can not be empty.


Siblings : Belong to same parent 

Level of a node: The count of edges on the path from the root node to that node. The root node has level 0.   

Depth:  Length of path from root node to the node for which we need to find the depth , Measure from  Root to node 

Height: In case of height we measure it from a node till bottom leaf node. 






```python
class Node:
    def __init__(self,data):
        self.data = data 
        self.right = None 
        self.left = None    
```

```python
def printTree(root):
    if root==None:
        return 
    print(root.data,end=": ")
    
    if root.left!=None:
        print("L",root.left.data,end=",")
    if root.right!=None:
        print("R",root.right.data,end="")

    print()
    printTree(root.left)
    printTree(root.right)
```
